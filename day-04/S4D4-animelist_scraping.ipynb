```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
mixed_reviews_url = "https://myanimelist.net/reviews.php?t=anime&p=1"

fmab_reviews_url = "https://myanimelist.net/anime/5114/Fullmetal_Alchemist__Brotherhood/reviews?p=1"

webpage = requests.get(fmab_reviews_url,
                       headers={"User-Agent": "Mozilla/5.0 (X11); Linux x86_64)"}).text
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
anime_reviews = soup.find_all("div", {"class": "borderDark"}) # hoverinfo_trigger
len(anime_reviews)
```

```python
columns = ["id", "title", "category", "date", "episodes_seen", "review", "reviewer"]
reviews = []

for review in anime_reviews:
    details = []
    
#     a_tag = review.find("a", {"class": "hoverinfo_trigger"})
#     anime_id = a_tag["id"].replace("#revArea", "") # getting id of anime for mixed review page
#     title = a_tag.get_text() # getting title of anime for mixed review page
    anime_id = 5114
    title = "Fullmetal Alchemist: Brotherhood"
    details.append(anime_id)
    details.append(title)
    
    category = review.find("small").get_text().replace("(", "").replace(")", "")
    details.append(category)
    
    divs = review.find_all("div")
    date = divs[2].get_text() + " " + divs[2]["title"]
    details.append(date)
    
    episodes_seen = review.find("div", {"class": "lightLink"}).get_text().replace("\n", "").replace(" episodes seen", "").strip()
    details.append(episodes_seen)
    
    text_info = review.find_all("div", {"class": ["spaceit", "textReadability"]})[3].get_text()
    text = text_info.strip()
    details.append(text)
    
    reviewer_info = review.find_all("td", {"valign": "top"})[1]
    reviewer = reviewer_info.find("a").get_text()
    details.append(reviewer)
    
    reviews.append(details)
    
reviews
```

```python
df = pd.DataFrame(reviews, columns=columns)
df
```
